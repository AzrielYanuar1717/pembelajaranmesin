Pemprocessing and Exploratory Data Analysis (EDA)

Menghapus Kolom dan Baris yang tidak dibutuhkan

```python
df = df.drop(['NO','TANGGAL','KODE','LABA','DEPOSIT PULSA','SALDO KEUANGAN'], axis=1)
df = df.drop(df.index[-1])
df
```

Mengubah Nama Kolom

```python
col_names = ['nama_barang','katagori','harga modal', 'harga jual','laris']
df.columns = col_names
df.columns
```

Mengecek Missing Value (NaN) 

```python
df.isnull().sum()
```

Mengisi Missing Value Menggunakan Forward fill

```python
df = df.fillna(method='ffill')
df.isnull().sum()
```

Mengkatagorikan kolom Harga Modal dan Harga Jual

Mengkatagorikan Harga Modal Menjadi:
. Rendah (<=25000)
. Sedang (>25000 & <=50000)
. Tinggi (>50000)

Mengkatagorikan Harga Jua; Menjadi:
. Rendah (<=27000)
. Sedang (>27000 & <=52000)
. Tinggi (>52000)

```python

modal = df['harga modal']
jual = df['harga jual']
df['harga modal'] = (np.where((modal <= 25000),'Rendah', np.where((modal <= 50000), 'Sedang', 'Tinggi'))).astype(str)
df['harga jual'] = (np.where((jual <= 27000),'Rendah', np.where((modal <= 52000), 'Sedang', 'Tinggi'))).astype(str)
df
```

Splitting

Memisahkan Data train dan test menggunakan train_test_split dengan perbandingan Train:test == 7:3

```python
from sklearn.model_selection import train_test_split
x = df.drop(['laris'], axis=1)
y = df['laris']
x_trial, x_test, y_trial, y_test = train_test_split(x,y,test_size=0.3, random_state = 0)
print(x_trial.shape)
print(x_test.shape)
print(x_trial)
print(y_trial)
```
